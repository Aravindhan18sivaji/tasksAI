```python
import pandas as pd
import ast  # To safely evaluate the dictionary-like string
import json


# Function to preprocess the column
def preprocess_product_details(value):
    try:
        # Convert the string into a valid dictionary format
        value = value.replace("=>", ":")  # Replace Ruby-style `=>` with Python's `:`

        # Convert to dictionary
        json_data = json.loads(value)

        # Extract product specifications
        specifications = json_data.get("product_specification", [])

        # Convert to a structured dictionary (key-value pairs)
        product_info = {}
        for item in specifications:
            key = item.get("key", None)
            val = item.get("value", None)
            if key and val:
                product_info[key] = val
            elif val:  # Some values have no "key"
                product_info[f"extra_{len(product_info) + 1}"] = val

        return product_info

    except Exception as e:
        return {"error": str(e)}

# Apply preprocessing to the DataFrame
df1['product_specifications'] = df1['product_specifications'].apply(preprocess_product_details)

# Show the result
# print(df["processed_product_details"])
```

```python
import ast  # To safely parse string lists

def preprocess_category_tree(value):
    try:
        # Convert string representation of list to actual list
        category_list = ast.literal_eval(value)  # Parses ["..."] safely

        if isinstance(category_list, list) and len(category_list) > 0:
            category_string = category_list[0]  # Extract the first category tree

            # Normalize separators, remove extra spaces, and convert to lowercase
            cleaned_category = category_string.replace(">>", "/").strip().lower()

            return cleaned_category  # Return as a clean string

    except Exception as e:
        return f"error: {e}"

# Apply preprocessing to the column
df2["product_category_tree"] = df2["product_category_tree"].apply(preprocess_category_tree)
```

```python
# import torch.nn.functional as F

class GraphSageItem:
    def __init__(self, A, X, Y, learning_rate, n_iterations, n_n_1, n_n_2, n_n_3):
        # Convert to torch tensors
        self.H = X
        self.Y = Y
        self.n_records = self.H.shape[0]
        self.n_features = self.H.shape[1]
        self.n_iterations = n_iterations
        self.learning_rate = learning_rate

        # Convert adjacency dictionary to PyTorch tensors
        self.A = {k: torch.tensor(list(v), dtype=torch.long) for k, v in A.items()}

        # Initialize Weights using Xavier Initialization
        self.W1 = torch.nn.Parameter(torch.empty(self.n_features, n_n_1))
        torch.nn.init.xavier_uniform_(self.W1)

        self.W2 = torch.nn.Parameter(torch.empty(n_n_1, n_n_2))
        torch.nn.init.xavier_uniform_(self.W2)

        self.W3 = torch.nn.Parameter(torch.empty(n_n_2, n_n_3))
        torch.nn.init.xavier_uniform_(self.W3)

        # Adam optimizer variables
        self.m1, self.v1 = torch.zeros_like(self.W1), torch.zeros_like(self.W1)
        self.m2, self.v2 = torch.zeros_like(self.W2), torch.zeros_like(self.W2)
        self.m3, self.v3 = torch.zeros_like(self.W3), torch.zeros_like(self.W3)

        self.beta1, self.beta2 = 0.9, 0.999  # Momentum coefficients
        self.epsilon = 1e-8  # Small constant to prevent division by zero
        self.t = 0  # Time step counter

    def mean_aggregate(self, H, sampled_neighbors):
        """Performs mean aggregation over the neighborhood."""
        # # print(H.dtype)
        # if not isinstance(H, torch.Tensor):
        #     H = torch.tensor(H, dtype=torch.float32)  # Convert to tensor if needed

        H_new = torch.zeros_like(H)
        for v, neighbors in sampled_neighbors.items():
            if len(neighbors) > 0:
                H_new[v] = torch.mean(H[neighbors], dim=0)
        return H_new




    def relu(self, H):
        return torch.where(H > 0, H, 0.01 * H)  # Leaky ReLU

    def relu_derivative(self, H):
        return torch.where(H > 0, torch.ones_like(H), 0.01 * torch.ones_like(H))

    def predictor(self, predict_in, predict_W):
        return predict_in @ predict_W  # Matrix multiplication in PyTorch

    def forward_propagation(self):
        # print("hi") # Ensure H is float
        A_H = self.mean_aggregate(self.H, self.A)
        self.H1 = A_H @ self.W1
        self.relu_H1 = self.relu(self.H1)

        A_H1 = self.mean_aggregate(self.relu_H1, self.A)
        self.H2 = A_H1 @ self.W2
        self.relu_H2 = self.relu(self.H2)
        self.y_cap = self.predictor(self.relu_H2, self.W3)



    def backward_propagation(self):
        for i in range(self.n_iterations):
            start_time = time.time()
            self.forward_propagation()

            # Compute gradients
            dL_dy_cap = (2 * (self.Y - self.y_cap)) / self.n_records
            dy_cap_dreluH2 = self.W3
            dreluH2_dH2 = self.relu_derivative(self.H2)
            dH2_dreluH1 = self.W2
            dreluH1_dH1 = self.relu_derivative(self.H1)
            dH1_dW1 = self.H

            dL_dW1 = (self.H.T @ ((((dL_dy_cap @ dy_cap_dreluH2.T) * dreluH2_dH2) @ dH2_dreluH1.T) * dreluH1_dH1)) / self.n_records
            dH2_dW2 = self.relu_H1
            dL_dW2 = dH2_dW2.T @ ((dL_dy_cap @ dy_cap_dreluH2.T) * dreluH2_dH2)
            dy_cap_dW3 = self.relu_H2
            dL_dW3 = dy_cap_dW3.T @ dL_dy_cap
            # Update weights using Adam
            self.update_weights(dL_dW1, dL_dW2, dL_dW3)
            self.loss = torch.mean((self.Y - self.y_cap) ** 2)
            epoch_end = time.time()
            epoch_time = epoch_end - start_time
            print(f"Epoch {i+1}/{self.n_iterations} - Time: {epoch_time:.4f} sec loss: {self.loss}")

    def update_weights(self, dL_dW1, dL_dW2, dL_dW3):
        """ Updates weights using the Adam optimizer. """
        self.t += 1  # Increment time step

        # **W1 Update**
        self.m1 = self.beta1 * self.m1 + (1 - self.beta1) * dL_dW1
        self.v1 = self.beta2 * self.v1 + (1 - self.beta2) * (dL_dW1 ** 2)
        m1_hat = self.m1 / (1 - self.beta1 ** self.t)
        v1_hat = self.v1 / (1 - self.beta2 ** self.t)

        # **W2 Update**
        self.m2 = self.beta1 * self.m2 + (1 - self.beta1) * dL_dW2
        self.v2 = self.beta2 * self.v2 + (1 - self.beta2) * (dL_dW2 ** 2)
        m2_hat = self.m2 / (1 - self.beta1 ** self.t)
        v2_hat = self.v2 / (1 - self.beta2 ** self.t)

        # **W3 Update**
        self.m3 = self.beta1 * self.m3 + (1 - self.beta1) * dL_dW3
        self.v3 = self.beta2 * self.v3 + (1 - self.beta2) * (dL_dW3 ** 2)
        m3_hat = self.m3 / (1 - self.beta1 ** self.t)
        v3_hat = self.v3 / (1 - self.beta2 ** self.t)

        # Perform weight updates without tracking gradients
        with torch.no_grad():
            self.W1 -= self.learning_rate * m1_hat / (torch.sqrt(v1_hat) + self.epsilon)
            self.W2 -= self.learning_rate * m2_hat / (torch.sqrt(v2_hat) + self.epsilon)
            self.W3 -= self.learning_rate * m3_hat / (torch.sqrt(v3_hat) + self.epsilon)


    def get_embedding(self):
        return self.relu_H2

# Convert data to PyTorch tensors
X = torch.tensor(X_train.values, dtype=torch.float32)
Y = torch.tensor(y_train.values, dtype=torch.float32)
Y = Y.unsqueeze(1)

# Convert adjacency graph to PyTorch tensor dictionary
A = {k: torch.tensor(list(v), dtype=torch.long) for k, v in graph.items()}

# Define neural network layer sizes
n_n_1 = int(X.shape[1])
n_n_2 = int(X.shape[1])
n_n_3 = 1

# Hyperparameters
learning_rate = 1
n_iterations = 5

# Create the GraphSAGE model
graphsageitem = GraphSageItem(A, X, Y, learning_rate, n_iterations, n_n_1, n_n_2, n_n_3)
```

```python
class Local_Sensitive_Hashing:
    def __init__(self,threshold,num_vectors):
        self.threshold = threshold
        self.num_vectors = num_vectors


    def cosine_similarity(self,item,vector):
        dot_product = np.dot(item,vector)
        normalized_a = np.linalg.norm(item)
        normalized_b = np.linalg.norm(vector)
        if normalized_a == 0 or normalized_b == 0:
            return 0
        return dot_product/(normalized_a * normalized_b)


    def separation(self):
        buckets = {}
        for item_index,item in enumerate(self.items):
            bucket_key = ""
            for vector in self.random_vector:
                similarity = self.cosine_similarity(item,vector)
                bucket_key += "1" if similarity > self.threshold else "0"
            if bucket_key not in buckets:
                buckets[bucket_key] = []
            buckets[bucket_key].append(item_index)
        return buckets


    def correctness_separation(self):
        correctness = {}
        for bucket_key,item_indices in self.buckets.items():
            correctness[bucket_key] = True
            for i in range(len(item_indices)):
                for j in range(i+1,(len(item_indices))):
                    similarity = self.cosine_similarity(self.items[item_indices[i]],self.items[item_indices[j]])
                    if similarity < self.threshold:
                        correctness[bucket_key] = False
                        break
                if not correctness[bucket_key]:
                    break
        return correctness

    def fit(self,items):
        num_columns = items.shape[1]
        self.items = items
        self.random_vector = np.random.randn(self.num_vectors,num_columns)
        self.buckets = self.separation()
        self.correctness = self.correctness_separation()

    def get_buckets(self):
        return self.buckets


    def get_correctness(self):
        return self.correctness
df = pd.read_csv('sample_user_preferences.csv')
df = df.values
threshold = 0.5
num_vectors = 4
lsh_model= Local_Sensitive_Hashing(threshold,num_vectors)
lsh_model.fit(df)
buckets = lsh_model.get_buckets()
correct = lsh_model.get_correctness()
```
