```python
import os
import json
from PIL import Image

# Define dataset path
dataset_path = "/root/.cache/kagglehub/datasets/marquis03/high-resolution-viton-zalando-dataset/versions/1/train"

# Define folders to load images from
image_folders = [
    "agnostic-v3.2", "cloth", "cloth-mask", "image",
    "image-densepose", "image-parse-agnostic-v3.2", "image-parse-v3",
    "openpose_img"
]

json_folder = "openpose_json"

# ✅ Define dataset dictionary BEFORE using it
dataset = {
    "image_paths": {},  # Store only file paths
    "json_data": {}  # Store parsed JSON content
}

# ✅ Populate dataset["image_paths"] to avoid NameError
for folder in image_folders:
    folder_path = os.path.join(dataset_path, folder)
    dataset["image_paths"][folder] = []  # Create list for each folder

    if os.path.exists(folder_path):
        for filename in os.listdir(folder_path):
            if filename.endswith((".jpg", ".png")):
                file_path = os.path.join(folder_path, filename)
                dataset["image_paths"][folder].append(file_path)  # Store path only

print(f"✅ Stored paths for images from {len(image_folders)} folders.")

# ✅ Load JSON files
json_folder_path = os.path.join(dataset_path, json_folder)
if os.path.exists(json_folder_path):
    for filename in os.listdir(json_folder_path):
        if filename.endswith(".json"):
            file_path = os.path.join(json_folder_path, filename)
            with open(file_path, "r") as f:
                dataset["json_data"][filename] = json.load(f)

print(f"✅ Loaded {len(dataset['json_data'])} JSON files.")
```

```python
import torch
import os
import numpy as np
import torch.nn.functional as F
from PIL import Image
from diffusers import StableDiffusionPipeline, UNet2DConditionModel, AutoencoderKL
from peft import LoraConfig, get_peft_model
import torchvision.transforms as transforms

# ✅ Define device
device = "cuda" if torch.cuda.is_available() else "cpu"

# ✅ Load Stable Diffusion model with VAE
# model_id = "CompVis/stable-diffusion-v1-4"
# pipeline = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to(device)
vae = AutoencoderKL.from_pretrained(model_id, subfolder="vae").to(device)

# ✅ Correct LoRA config for UNet
config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["conv1", "conv2", "conv_shortcut"],  # Only target Conv2d layers
    lora_dropout=0.1
)

# ✅ Apply LoRA to UNet and move it to CUDA
pipeline.unet = get_peft_model(pipeline.unet, config).to(device)
pipeline.unet.print_trainable_parameters()

# ✅ Ensure Text Encoder is on CUDA
pipeline.text_encoder.to(device)

# ✅ Move tokenizer to CUDA
# pipeline.tokenizer.to(device)

# ✅ Move optimizer parameters to CUDA
optimizer = torch.optim.AdamW(pipeline.unet.parameters(), lr=1e-5)

# ✅ Define dataset path
train_pairs = list(zip(
    dataset["image_paths"].get("agnostic-v3.2", []),
    dataset["image_paths"].get("cloth", [])
))

# ✅ Image transformation (normalize + convert format)
transform = transforms.Compose([
    transforms.Resize((512, 512)),   # Resize to match model input
    transforms.ToTensor(),           # Convert to (C, H, W) format
    transforms.Normalize([0.5], [0.5])  # Normalize to [-1,1] (Stable Diffusion requires this)
])

# ✅ Function to encode images
def encode_image(image_path):
    image = Image.open(image_path).convert("RGB")
    image_tensor = transform(image).unsqueeze(0).to(device)  # ✅ Move image to CUDA

    # ✅ Pass through VAE encoder (ensure VAE is on CUDA)
    with torch.no_grad():
        latent = vae.encode(image_tensor).latent_dist.sample().to(dtype=torch.float16) * 0.18215


    return latent  # Shape: (1, 4, 64, 64)

# ✅ Define timestep and move to CUDA
timestep = torch.tensor([10], device=device, dtype=torch.long).to(dtype=torch.float16)
  # Example timestep

# ✅ Training loop
for epoch in range(3):  # Train for 3 epochs
    for human_path, cloth_path in train_pairs:
        openpose_path = human_path.replace("agnostic-v3.2", "openpose_img")
        openpose_path = openpose_path.replace(".jpg", "_rendered.png")

        if not (os.path.exists(human_path) and os.path.exists(cloth_path) and os.path.exists(openpose_path)):
            print(f"Skipping missing files: {human_path}, {cloth_path}, {openpose_path}")
            continue  # Skip missing files

        # ✅ Encode images and move them to CUDA
        human_latent = encode_image(human_path).to(device)
        cloth_latent = encode_image(cloth_path).to(device)
        openpose_latent = encode_image(openpose_path).to(device)

        # ✅ Create conditioning input and move to CUDA
        conditioning_input = (human_latent + cloth_latent + openpose_latent) / 3



        # ✅ Define training prompt
        prompt = "Generate an image of this exact person, preserving face and body shape, wearing the given clothing item naturally."

        # ✅ Get text embeddings for conditioning (ensure everything is on CUDA)
        input_ids = pipeline.tokenizer(prompt, return_tensors="pt").input_ids.to(device)
        text_embeddings = pipeline.text_encoder(input_ids)[0].to(device, dtype=torch.float16)


        # ✅ Forward pass through U-Net
        optimizer.zero_grad()
        pred_latents = pipeline.unet(conditioning_input, timestep, encoder_hidden_states=text_embeddings).sample

        # ✅ Compute loss (MSE)
        loss = F.mse_loss(pred_latents, human_latent)  # Compare with original human latent
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch+1} completed, Loss: {loss.item()}")

# ✅ Save fine-tuned model
pipeline.save_pretrained("./fine-tuned-stable-diffusion-vton")
print("Model fine-tuned and saved successfully!")
```
