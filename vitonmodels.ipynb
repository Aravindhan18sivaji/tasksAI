```python
from diffusers import ControlNetModel, StableDiffusionControlNetPipeline

controlnet = ControlNetModel.from_pretrained("hoa-quickloop/tryon_controlnet")
pipe = StableDiffusionControlNetPipeline.from_pretrained(
	"stabilityai/stable-diffusion-2-1-base", controlnet=controlnet
)
```

```python
from PIL import Image
```

```python
# Load input images
person_image = Image.open("/content/t-shirt.jpeg").convert("RGB")
cloth_image = Image.open("/content/image-wear.jpeg").convert("RGB")

# Define a try-on prompt
prompt = "Generate an image of the given person wearing the given cloth."

# Generate the try-on result
result = pipe(prompt=prompt, image=person_image, clothing_image=cloth_image).images[0]

# Display the result
result.show()

# Save the generated image
result.save("output_tryon.jpg")
```
